```python
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Lasso, ElasticNet, Ridge
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score 
import numpy as np 
```

```python
bank = pd.read_csv(r"..\Cases\Bankruptcy\Bankruptcy.csv")

y = bank['D']
X = bank.drop(['NO', 'D'], axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, StratifiedKFold


lr = LogisticRegression(solver='saga')
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=24)

params = {'penalty':['elasticnet', 'L1', 'L2', None],
          'C':np.linspace(0.001, 10, 5),
          'l1_ratio':np.linspace(0.001, 1, 4)
          }

gcv = GridSearchCV(lr, param_grid=params, cv=kfold,scoring='neg_log_loss')
```

```python
gcv.fit(X,y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
pd_cv = pd.DataFrame(gcv.cv_results_)
```
