```python
import pandas as pd
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score, KFold
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score, log_loss, roc_auc_score, r2_score
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB, BernoulliNB
import numpy as np
```

```python
df = pd.read_csv(r'..\Cases\Glass Identification\Glass.csv')
df
```

```python

le = LabelEncoder()
y = le.fit_transform(df['Type'])
X = df.drop('Type', axis=1)
print(le.classes_)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=24, stratify=y)
```

```python
lg = LogisticRegression()
lg.fit(X_train,y_train)
y_pred = lg.predict(X_test)
y_pred_prob = lg.predict_proba(X_test)

print(accuracy_score(y_test, y_pred))
print(roc_auc_score(y_test, y_pred_prob, multi_class='ovr'))
print(log_loss(y_test, y_pred_prob))
```
