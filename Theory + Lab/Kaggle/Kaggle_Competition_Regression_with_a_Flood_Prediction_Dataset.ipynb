```python
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import log_loss
```

```python
train = pd.read_csv(r"train.csv")
train.head()
```

```python
test = pd.read_csv(r"test.csv")
test.head()
```

```python
X_train = train.drop(["FloodProbability", "id"], axis = 1 )
```

```python
y_train = train["FloodProbability"]
```

```python
X_test = test.drop("id", axis = 1 )
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred_prob = lr.predict(X_test)
```

```python
submit = pd.DataFrame({"id":test['id'],"FloodProbability":y_pred_prob})
```

```python
submit.to_csv("sbt_lr.csv" , index = False)
```

```python
from sklearn.linear_model import ElasticNet


elastic = ElasticNet(alpha=4)

elastic.fit(X_train,y_train)
print(elastic.coef_)
print(elastic.intercept_)
elastic.predict(X_test)
```

```python
elastic = ElasticNet()
import numpy as np
from sklearn.model_selection import GridSearchCV, KFold
kfold = KFold(n_splits=5, shuffle=True, 
              random_state=24)

print(elastic.get_params())
params = {'alpha':np.linspace(0.001, 50,5),
          'l1_ratio':np.linspace(0.001,1, 10)}
gcv = GridSearchCV(elastic, param_grid=params,
                   cv=kfold, scoring='r2')
gcv.fit(X_train, y_train)
pd_cv = pd.DataFrame(gcv.cv_results_)
print(gcv.best_params_)
print(gcv.best_score_)

```
