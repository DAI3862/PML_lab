### Bike Sharing

```python
df = pd.read_csv(r'..\Kaggle\Bike Price\train.csv', parse_dates= ['datetime'])
df.head()
```

```python
df.set_index('datetime', inplace = True)
```

```python
casual = df['casual']
mon_cou = casual.resample('M').sum()
mon_cou.index.rename('Month', inplace=True)
mon_cou.plot()

plt.title('Monthly Causal Rentals')
plt.xlabel('Months')
plt.show()
```

```python
mon_cou = casual.resample('Q').sum()
mon_cou.index.rename('Quater', inplace=True)
mon_cou.plot()

plt.title('Quaterly Causal Rentals')
plt.xlabel('Quaterly')
plt.show()
```

```python
registered = df['registered']
mon_cou = registered.resample('M').sum()
mon_cou.index.rename('Month', inplace=True)
mon_cou.plot()

plt.title('Monthly registered')
plt.xlabel('Months')
plt.show()
```

```python
mon_cou = registered.resample('Q').sum()
mon_cou.index.rename('Quater', inplace=True)
mon_cou.plot()

plt.title('Quaterly registered')
plt.xlabel('Quaterly')
plt.show()
```
